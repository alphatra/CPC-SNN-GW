```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, Dropdown, Output, VBox, HBox
from IPython.display import display

# >>> TU WPISZ SWOJĄ ŚCIEŻKĘ DO PLIKU <<<
H5_PATH = "../data/cpc_snn_train.h5"

h5 = h5py.File(H5_PATH, "r")

print("Załadowano plik:", H5_PATH)
print("Struktura najwyższego poziomu:")
for key in h5.keys():
    print(" -", key)
```

```python
def list_children(group):
    """Zwraca listę dzieci (grup/datasetów) w danej grupie."""
    return list(group.keys())

def describe_item(item):
    """Zwraca opis elementu: typ, shape, dtype, atrybuty."""
    info = {}
    if isinstance(item, h5py.Dataset):
        info['type'] = 'dataset'
        info['shape'] = item.shape
        info['dtype'] = str(item.dtype)
    else:
        info['type'] = 'group'

    info['attributes'] = {k: v for k, v in item.attrs.items()}
    return info
```

```python
events = sorted(list(h5.keys()))
event_dropdown = Dropdown(options=events, description="Event:")
```

```python
output = Output()

def show_event(event_name):
    output.clear_output()
    with output:
        grp = h5[event_name]

        print(f"▶ EVENT: {event_name}")
        print("Zawartość:")
        for key in grp.keys():
            print("  -", key)

        print("\nAtrybuty eventu:")
        for k, v in grp.attrs.items():
            print(f"  {k} = {v}")

        # Jeśli dostępny dataset "H1" lub "L1" – pokaż sygnał
        for det in ["H1", "L1"]:
            if det in grp:
                ds = grp[det]
                if isinstance(ds, h5py.Dataset):
                    print(f"\nWykres: {det}")
                    plt.figure(figsize=(12,3))
                    plt.plot(ds[:])
                    plt.title(f"{event_name} – {det}")
                    plt.xlabel("sample")
                    plt.ylabel("strain")
                    plt.show()

interact(show_event, event_name=event_dropdown)
display(output)
```

```python
def explore_group(event_name, path):
    grp = h5[event_name][path]
    print("Zawartość grupy:", path)
    for key in grp.keys():
        print(" -", key)
```
