# CPC-SNN Training Analysis

This notebook provides an interactive dashboard to visualize training metrics from Weights & Biases.

```python
%matplotlib inline
import sys
import os

# Add project root to path
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
if project_root not in sys.path:
    sys.path.append(project_root)

from src.analysis.wandb_plotter import WandBPlotter

# Initialize Plotter
plotter = WandBPlotter(entity="alpphatra", project="cpc-snn-gw")
```

## Recent Runs

```python
runs = plotter.get_runs(limit=5)
for r in runs:
    print(f"Run: {r.name} (ID: {r.id}) - Status: {r.state}")
```

## Single Run Analysis
Enter the Run ID below to visualize its metrics.

```python
# Replace with your Run ID
run_id = runs[0].id if runs else ""
print(f"Analyzing Run: {run_id}")

# Standard Metrics
plotter.plot_run_metrics(run_id, smooth=0.0)
```

## Advanced SNN Metrics
Spike density, Context statistics, etc.

```python
plotter.plot_advanced_metrics(run_id)
```

## Compare Runs
Compare validation loss across multiple runs.

```python
if len(runs) >= 2:
    run_ids = [r.id for r in runs[:3]]
    plotter.compare_runs(run_ids, metric="val_loss", smooth=0.0)
else:
    print("Not enough runs to compare.")
```
